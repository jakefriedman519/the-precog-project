```python
state_abbrev = pd.read_csv('../../../../static/datasets/state_abbrev.csv', index_col=0)
state_abbrev.loc['Alabama', 'Abbreviation']
```

```python
age_data = pd.read_csv('../../../../static/datasets/uscensus/age_distribution.csv', index_col=0).drop('Footnotes', axis=1)
age_ranges = ['0-18', '19-25', '26-34', '35-54', '55-64', '65+']
def age_to_range(age):
    if age == 'Unknown' or np.isnan(age) or age is None:
        return 'Unknown'
    age = int(age)
    if age < 19:
        return '0-18'
    elif age < 26:
        return '19-25'
    elif age < 35:
        return '26-34'
    elif age < 55:
        return '35-54'
    elif age < 65:
        return '55-64'
    else:
        return '65+'
        
age_data.columns = [*age_ranges, 'total']
age_data.index = state_abbrev.loc[age_data.index, 'Abbreviation'].values
age_data.head()
```

```python
census_data = pd.read_csv('../../../../static/datasets/uscensus/acs2015_census_tract_data.txt', index_col=0).dropna(axis=0)
census_data['state'] = state_abbrev.loc[census_data['state'].values, 'Abbreviation'].values
census_data.head()
```

```python
census_data['menRatio'] = census_data['men'] / census_data['totalpop']
census_data['womenRatio'] = census_data['women'] / census_data['totalpop']
df = census_data[['state', 'totalpop', 'menRatio', 'womenRatio', 'hispanic', 'white', 'black', 'native', 'asian', 'pacific']]
df = df.groupby('state').apply(lambda x: pd.DataFrame(index=[0], data=dict(zip(map(str, x.drop('totalpop', axis=1).columns.to_list()), np.average(x.drop('totalpop',axis=1), weights=x.totalpop, axis=0))))).rename(columns={'menRatio': 'male', 'womenRatio': 'female', 'hispanic': 'H', 'white': 'W', 'black': 'B', 'native': 'N', 'asian': 'A', 'pacific': 'P'}).reset_index(level=1, drop=True)
df = pd.concat([census_data[['state', 'totalpop']].groupby('state').sum(), df], axis=1)
df['male'] = df['male'] * 100
df['female'] = df['female'] * 100
df = df.drop('PR', axis=0)
df = df.sort_index()
for age_range in age_ranges:
    df[f'{age_range}_ratio'] = age_data.sort_index().loc[:, age_range].values / age_data.sort_index().loc[:, 'total'].values
df.to_csv('../../../../static/datasets/uscensus/acs2015_census_tract_data_aggregated.csv')
df
```

```python
homicide_data = pd.read_csv('../../../../static/datasets/washingtonpost/homicide-data.csv', index_col=0)
homicide_data = homicide_data.drop(['victim_last', 'victim_first', 'lat', 'lon'], axis=1)
homicide_data['victim_race'] = homicide_data['victim_race'].apply(lambda x: x[0])
homicide_data['victim_age_range'] = homicide_data['victim_age'].apply(age_to_range)
homicide_data['reported_date'] = homicide_data['reported_date'].apply(lambda x: int(str(x)[:4]))
homicide_data['victim_sex'] = homicide_data['victim_sex'].apply(lambda x: x.lower())
homicide_data.columns = ['year', 'race', 'age', 'gender', 'city', 'state', 'outcome', 'age_range']
homicide_data = homicide_data.reset_index(drop=True)
homicide_data.to_csv('../../../../static/datasets/washingtonpost/homicide-data-augmented.csv')
homicide_data
```
